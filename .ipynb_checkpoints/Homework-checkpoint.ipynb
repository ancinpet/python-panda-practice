```python
import pandas
```

```python
pandas.set_option('display.max_columns', None)
pandas.set_option('min_rows', 50)
survey = pandas.read_csv('data/survey_results_public.csv')
survey
```

```python
people_total = len(survey)
print('1. Kolik lidí celkem vyplnilo dotazník?')
print(f'Dotazník vyplnilo: {people_total} lidí.')
```

```python
our_people = survey[survey['Country'].str.contains('Czech|Slovakia', na=False)]
our_people_total = len(our_people)
print('2. Kolik z nich bylo z ČR/SR?')
print(f'Dotazník vyplnilo: {our_people_total} lidí z ČR/SR.')
```

```python
def wage_of_survey(survey, usd_to_czk):
    wage_monthly = survey['ConvertedComp'] * usd_to_czk / 12.0
    return wage_monthly.mean(skipna=True)

def get_usd_to_czk(survey):
    wanted = survey[(survey['CurrencySymbol'].str.contains('CZK', na=False)) & (survey['CompFreq'].str.contains('Yearly', na=False))]
    wanted_row = wanted.loc[wanted.first_valid_index()]
    return wanted_row['CompTotal'] / wanted_row['ConvertedComp']

print('3. Jaká je, podle dotazníku, průměrná mzda programátorů v ČR/SR? Ve světě? Odpovídejte v Kč měsíčně.')
usd_to_czk = get_usd_to_czk(survey)
wage_our = wage_of_survey(our_people, usd_to_czk)
wage_world = wage_of_survey(survey, usd_to_czk)
print(f'Průměrná mzda programátorů v ČR/SR je {round(wage_our, 2)} Kč měsíčně.')
print(f'Průměrná mzda programátorů ve světě je {round(wage_world, 2)} Kč měsíčně.')
```

```python
def uses_language(survey, language):
    return survey[survey['LanguageWorkedWith'].str.contains(f'{language}$|{language};', na=False)]

def doesnt_use_language(survey, language):
    return survey[~(survey['LanguageWorkedWith'].str.contains(f'{language}$|{language};', na=False))]

def wants_to_use_language(survey, language):
    return survey[survey['LanguageDesireNextYear'].str.contains(f'{language}$|{language};', na=False)]

print('4. Kolik lidí z ČR/SR uvedlo, že používá Python?')
our_python_users = uses_language(our_people, 'Python')
print(f'Z ČR/SR používá Python {len(our_python_users)} lidí.')

print('')

print('5. Kolik z nich chce v používání Pythonu pokračovat?')
our_continue_python = wants_to_use_language(our_python_users, 'Python')
print(f'V používání Pythonu chce v ČR/SR pokračovat {len(our_continue_python)} lidí.')

print('')

print('6. Kolik lidí Python nepoužívá, ale chce ho začít používat?')
world_doesnt_use_python = doesnt_use_language(survey, 'Python')
world_wants_to_start_python = wants_to_use_language(world_doesnt_use_python, 'Python')
print(f'Python celosvětově nepoužívá {len(world_doesnt_use_python)} lidí, ale chce ho začít používat {len(world_wants_to_start_python)} lidí z nich.')
```

```python
print('Předcházející tři otázky odpovězte také pro Javu místo Pythonu.')
print('')

print('7a. Kolik lidí z ČR/SR uvedlo, že používá Javu?')
our_java_users = uses_language(our_people, 'Java')
print(f'Z ČR/SR používá Javu {len(our_java_users)} lidí.')

print('')

print('7b. Kolik z nich chce v používání Javy pokračovat?')
our_continue_java = wants_to_use_language(our_java_users, 'Java')
print(f'V používání Javy chce v ČR/SR pokračovat {len(our_continue_java)} lidí.')

print('')

print('7c. Kolik lidí Javu nepoužívá, ale chce ji začít používat?')
world_doesnt_use_java = doesnt_use_language(survey, 'Java')
world_wants_to_start_java = wants_to_use_language(world_doesnt_use_java, 'Java')
print(f'Javu celosvětově nepoužívá {len(world_doesnt_use_java)} lidí, ale chce ji začít používat {len(world_wants_to_start_java)} lidí z nich.')
```

```python
print('8. Vykreslete graf, který pro každý programovací jazyk, který lidi z ČR/SR uvedli, ukáže průměrnou měsíční mzdu lidí z ČR/SR, kteří tento jazyk používají (podle dat dostupných z dotazníku).')
exploded_by_langauge = \
    (our_people.set_index(our_people.columns.drop('LanguageWorkedWith',1).tolist())
    .LanguageWorkedWith.str.split(';', expand=True)
    .stack()
    .reset_index()
    .rename(columns={0:'LanguageWorkedWith'})
    .loc[:, our_people.columns])
languages = exploded_by_langauge.groupby(['LanguageWorkedWith']).mean()
languages = languages.filter(['ConvertedComp'])
languages['ConvertedComp'] = languages['ConvertedComp'] * usd_to_czk / 12
graph = languages['ConvertedComp'].plot.bar(title='Průměrná měsiční mzda lidí z ČR/SR pro jednotlivé jazyky', xlabel='Programovací jazyky', ylabel='Mzda měsičně (CZK)')
```
